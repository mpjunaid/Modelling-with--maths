```python
def getmonth(x):
    return int(x['daily sale'].month)
```

```python
a = [(0, 1),(0, 4),(0, 6),(0, 8),(0, 9),(0, 11),(1, 5),(2, 0),(3, 0),(4, 7),\
     (5, 0),(6, 3),(7, 0),(8, 2),(9, 0),(10, 0),(11, 10)]
b = [1,3,4,6,7,8,11,12,14,17,18]

def groupy(a,b):
    
    b.insert(0,0)
    
    flag = [False]*len(a)
    res = []
    i = 0

    while(i < len(a)):
        if flag[i]:
            i+=1
        elif a[i][0]==0:
            flag[i] = True
            inp = []
            j = i+1
            target = a[i][1]

            while j < len(a):
                if (flag[j]) or (a[j][0]!=target):
                    j+=1
                elif a[j][0]==target:
                    flag[j]=True
                    inp.append(b[a[j][0]])
                    target=a[j][1]
                    j=i+1
                    if target==0:
                        break
                else:
                    j+=1
            i+=1
            res.append(inp)
 
    return res

groupy(a,b)
```
